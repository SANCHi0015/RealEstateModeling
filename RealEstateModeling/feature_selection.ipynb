```python
def categorize_floor(floor):
    if 0 <= floor <= 2:
        return "Low Floor"
    elif 3 <= floor <= 10:
        return "Mid Floor"
    elif 11 <= floor <= 51:
        return "High Floor"
    else:
        return None
```

```python
df.floorNum = df.floorNum.apply(categorize_floor)
```

```python
df.rename(columns={
    'floorNum' : 'floor'
},inplace=True)
```

**Convert all categorical col in numerical usin ordinal encoding**

- why ordinal encoding
- Preserve Order Information:
  If your categorical data has a natural order (e.g., "low," "medium," "high"), ordinal encoding converts these into numbers that reflect this order (e.g., 1, 2, 3). This way, the model can recognize that "high" is greater than "medium," and "medium" is greater than "low."

- compatibility with ml algorithms 
some algo works better with numeric values

- Why Use Double Brackets Here?
- When applying OrdinalEncoder (or many other sklearn transformers), the input is expected to be a DataFrame (2D array) rather than a Series (1D array). Using double brackets ensures that the selected column is in the correct format for the transformation.

```python
from sklearn.preprocessing import OrdinalEncoder

data = df.copy()

categorical_col = df.select_dtypes(include=['object']).columns

oe = OrdinalEncoder()
for col in categorical_col:
    data[col] = oe.fit_transform(data[[col]])

x = data.drop('price' , axis=1)
y = data.price
```

# ***Corrrelation Analysis***

```python
corr = data.corr()['price']
corr = corr.drop('price').reset_index().rename(columns={
    'index' : 'features',
    'price' : 'corr'
})
corr
```

# ***Randomforest feature Analysis***

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=100 , random_state=42)
rf.fit(x,y)

rfa = pd.DataFrame({
    'features' : x.columns,
    'rf' : rf.feature_importances_
})
rfa
```

# ***Gradient Boosting Analysis***

```python
from sklearn.ensemble import GradientBoostingRegressor

gb = GradientBoostingRegressor()
gb.fit(x,y)

gba = pd.DataFrame({
    'features' : x.columns,
    'gb' : gb.feature_importances_
})
gba
```
