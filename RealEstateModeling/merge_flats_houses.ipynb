```python
import pandas as pd
```

```python
houses = pd.read_csv("houses_cleaned.csv" , index_col=[0])
flats = pd.read_csv("flats_cleaned.csv" , index_col=[0])
```

```python
houses.rename(columns={"noOfFloor":"floorNum"} , inplace=True)
```

```python
df = pd.concat([flats, houses] , ignore_index=True)
```

```python
df = df.sample(df.shape[0], ignore_index=True)
```

```python
df.to_csv("gurgaon_properties_data.csv")
```
