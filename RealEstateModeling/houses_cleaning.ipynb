```python
import pandas as pd
import numpy as np
import re
```

```python
houses = pd.read_csv("houses.csv")
```

```python
houses.sample(5)
```

```python
houses.drop(columns=["link" , "property_id"] , inplace=True)
```

```python
houses = houses[houses.price != "Price on Request"]
```

```python
def lac(x):
    if (type(x) == float):
        return x
    else:
        if(x[1] != "Lac"):
            return round (float(x[0]) , 2)
        else:
            return round(float(x[0])/100,2)
```

```python
houses.price = houses.price.str.split(" ").apply(lac)
```

```python
houses.rename(columns={"rate" : "price_per_sq"},inplace=True)
```

```python
houses.price_per_sq = houses.price_per_sq.str.replace(",", "").str.replace("₹ " , "").str.replace("/sq.ft." , "").astype(float)
```

```python
houses = houses[~houses.bedRoom.isnull()]
```

```python
houses['bedRoom'] = houses.bedRoom.str.replace("Bedrooms" , "").str.replace("Bedroom","")
```

```python
houses.bedRoom = houses['bedRoom'].astype(int)
```

```python
houses['bathroom'] = houses.bathroom.str.replace("Bathrooms" , "").str.replace("Bathroom","")
```

```python
houses.bathroom = houses.bathroom.astype(int)
```

```python
houses.balcony = houses.balcony.str.replace("+" , "").str.replace(" Balconies","").str.replace(" Balcony", "").str.replace("No ", "0")
```

```python
houses.balcony = houses.balcony.str.replace("No","0")
```

```python
houses.balcony = houses.balcony.astype(int)
```

```python
houses.society = houses.society.apply (lambda x : re.sub(r"\d+\.+\d+\s★","" , str(x))).str.strip().str.lower()
```

```python
houses.additionalRoom.fillna("not available" , inplace=True)
```

```python
houses.additionalRoom = houses.additionalRoom.str.lower()
```

```python
houses.address.fillna("Na" , inplace=True)
```

```python
houses.head()
```

```python
houses.noOfFloor = (
    houses.noOfFloor.str.split(" ").str.get(0)
    .str.replace("Basement" , "-1")
    .str.replace("Ground","0")
    .str.replace("Lower" , "0").str.extract(r"(\d+)")
)
```

```python
houses.facing.fillna("Na",inplace=True)
```

```python
houses.nearbyLocations = houses.nearbyLocations.str.lower()
```

```python
houses.furnishDetails = houses.furnishDetails.str.lower()
```

```python
houses.features = houses.features.str.lower()
```

```python
houses.rating = houses.rating.str.lower()
```

```python
houses.area = round((houses.price * 10000000) / houses.price_per_sq)
```

```python
houses.insert(column="property_type" , value="houses",loc=2)
```

```python
houses.society = houses.society.str.replace("nan" , "independent")
```

```python
houses.to_csv("houses_cleaned.csv")
```
