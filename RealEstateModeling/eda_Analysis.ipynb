```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('gurgaon_v4.csv')
```

```python
df.describe()
```

```python
df.info()
```

```python
df
```

```python
plt.figure(figsize=[15,6])

plt.subplot(1,2,1)
sns.histplot(x=df.price  , kde=True,bins=50, color='maroon')
plt.title('Distribution of price column')
plt.xlabel('price')
plt.ylabel('frequency')
plt.grid()

plt.subplot(1,2,2)
sns.histplot(x=np.log1p(df.price) ,bins=50, kde=True,color='hotpink')
plt.title('boxplot of log price')
plt.xlabel('price')
plt.ylabel('frequency')
plt.grid()

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=[15,6])

plt.subplot(1,2,1)
sns.boxplot(x=df.price , color='maroon')
plt.title('Distribution of price column')
plt.xlabel('price')
plt.ylabel('frequency')
plt.grid()

plt.subplot(1,2,2)
sns.boxplot(x=np.log1p(df.price) ,color='pink')
plt.title('boxplot of log price')
plt.xlabel('price')
plt.ylabel('frequency')
plt.grid()

plt.tight_layout()
plt.show()
```
